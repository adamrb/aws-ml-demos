```python
import os, boto3

polly = boto3.client('polly')

def speak(polly, text, text_type="text", format='mp3', voice='Brian'):
    file_path = os.path.join(os.curdir, "sound.mp3")
    if os.path.exists(file_path):
        os.remove(file_path)
    resp = polly.synthesize_speech(OutputFormat=format, Text=text, TextType=text_type, VoiceId=voice)
    soundfile = open(file_path, 'wb')
    soundBytes = resp['AudioStream'].read()
    soundfile.write(soundBytes)
    soundfile.close()
```

```python
speak(polly, 
      "Hello world, I'm Polly. Or Brian. Or anyone you want, really.",
     text_type="text")
```
